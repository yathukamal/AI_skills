```python
#Import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import seaborn as sns
```

```python
#read in titanic file
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
titanic.plot()
plt.show()
```

```python
#line plot
titanic.plot.line(x='Age',y='Fare',figsize=(8,6))
```

```python
#scatterplot
titanic.plot.scatter(x='Age',y='Fare',figsize=(8,6))
```

```python
#Box Plot
titanic.plot.box(figsize=(10,8))
```

```python
#Hexagonal Plot
titanic.plot.hexbin(x='Age',y='Fare',gridsize=30,figsize=(8,6))
```

```python
#Kernel Density Plot
titanic['Age'].plot.kde()
```

```python
#Bar Plots
titanic['Pclass'].value_counts().sort_index().plot.bar()
```

```python
#Horizontal bar plot
titanic['Survived'].value_counts().sort_index().plot.barh()
```

```python
titanic['Pclass'].plot.bar()
```

```python
titanic['Age'].hist()
```

```python
titanic['Age'].hist(bins=5)
```

```python
#Change style of the grid
sns.set_style("darkgrid")
```

```python
#Plotting histogram again to see the colour change
titanic['Age'].hist(bins=20)
```

```python
#Plotting multiple attributes
plt.figure(figsize=(10,6))
sns.set_style("whitegrid")
ax = sns.countplot(x='Survived',hue = 'Sex', data=titanic, palette='Greys')
plt.xticks(rotation = 0, ha='right')
plt.savefig('titanic_countplot.png')
```
