```python
#How many missing values are in the dataset?
column_names = [
    "age",
    "workclass",
    "fnlwgt",
    "education",
    "education-num",
    "marital-status",
    "occupation",
    "relationship",
    "race",
    "sex",
    "capital-gain",
    "capital-loss",
    "hours-per-week",
    "country",
    "salary",
]

adult=pd.read_csv("adult.csv", names=column_names , na_values="?",skipinitialspace=True)
adult.isnull().sum()
```

```python
#Calculate the mean of the hours an individual has reported working per week presented in the dataset
#Know that the dtype is an object so we need to change to numeric
adult["hours-per-week"]=pd.to_numeric(adult["hours-per-week"], errors="coerce")
adult["hours-per-week"].mean()

#40.44 hours
```

```python
#Using the count function, how many male and female individuals
#drop the first row
adult = adult.iloc[1:].reset_index(drop=True)
adult["sex"].value_counts()
```

```python
#Plot the "work class" attribute using the bar plot and value counts function
adult["workclass"].value_counts().sort_index().plot.bar()
```

```python
#Using a histogram plot, visualise the age range
adult["age"].hist()
```

```python
#Visualise the relation between two attributes using the bar plots
sns.barplot(x='sex',y='salary',data=adult)
plt.xticks(rotation=0,ha='right')
plt.savefig('adult_barplot.png')
```
